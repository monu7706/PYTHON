## oops(object oriented programming)

```python
class car:
    ## attrobutes
    def car_data(self,color,brand):
        self.color=color
        self.brand=brand
    ## methods
    def start(self):
        print(f"the car of this {self.color} and brand of {self.brand} has started")

car1=car()
car1.car_data("black","mg")
car1.start()
```

```python
class car:
    ## attrobutes
    def __init__(self,color,brand,switches):
        self.color=color
        self.brand=brand
        self.switches=switches
    ## methods
    def start(self):
        print(f"car will start on clicking by clicking no. of switches {self.switches}")
        print(f"the car of this {self.color} and brand of {self.brand} has started")

car1=car("black","mg",5)
car1.start()
```

```python
class bank_account:
    def __init__(self,acc_holder_name,balance):
        self.acc_holder_name=acc_holder_name
        self.balance=balance

    def deposit(self,amount):
        self.balance=self.balance+amount
        print(f"you have made successful deposit your balance {self.balance}")

    def withdrawl(self,amount):
        if amount<=self.balance:
            self.balance=self.balance-amount
            print(f"you have successful withawl remaining balance{self.balance}")
        else:
            print("insufficient")
acc1=bank_account("adarsh",5000)
acc1.deposit(2000)
acc1.withdrawl(3000)

```

```python
class car:
    ## attrobutes
    def __init__(self,color,brand,):
        self.color=color
        self.brand=brand
    ## methods

    def start(self,switches):
        self.switches=switches
        print(f"car will start on clicking by clicking no. of switches {switches}")
        print(f"the car of this color {self.color} and brand of {self.brand} has started")
        
    def show(self):
        print(f"no of switches are {self.switches}")

car1=car("black","mg")
car1.start(5)
car1.show()
```

## 1. Encapsulation

```python
class Bank_balance:
    def __init__(self, owner,__balance):
        self.owner=owner
        self.__balance=__balance    ## " __ " private kr sakte hai
b=Bank_balance("ritik",5000)
print(b.__balance)                   ## __ se pehle ek _ aur class ka name paste then dekh sakte hai
print(b.owner)
b.__balance + 2000               ## bank._bank_balance__balance
```

```python
class Bankbalance:
    def __init__(self, owner,__balance):
        self.owner=owner
        self.__balance=__balance
    def get(self):
        print(f"the total balance remaining is {self.__balance}")
    def set(self,amount):
        self.amount=amount
        if self.amount>0:
            self.__balance=self.__balance+amount
        else:
            print("the amount is null")
        
            
bank=Bankbalance("ritik",5000)
bank.set(5000)
bank.get()
          
```

## 2. inheritance

```python
class Bus:
    def __init__(self,color,brand,model):
        self.color=color
        self.brand=brand
        self.model=model
    def show (self):
        print(f"I have bus of color {self.color} and brand of {self.brand} amd model {self.model}")
    
class electric_bus(Bus):
    def __init__(self,color,brand,model,battery_capacity):
        super(). __init__(color,brand,model)
        self.battery_capacity = battery_capacity
    def show(self):
        print(f"I have bus of color {self.color} brand {self.brand} and model {self.model} and battery_capacity {self.battery_capacity}")

class cng_bus(Bus):
    def __init__(self,color,brand,model, milage):
        super(). __init__(color,brand,model)
        self.milage = milage
    def show (self):
        print(f"I have cng bus color {self.color} brand {self.brand} model {self.model} and of milage {self.milage}")

b1=Bus("red","volvo",2017)
b1.show()
b2=electric_bus("blue","mg","2013","5000mah")
b2.show()
b3=cng_bus("pink","audi",2024,"20kmpl")
b3.show()
```
