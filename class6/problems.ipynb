## problem no. 2
### ** factorial **
### i.e 
### 1! = 1*1
### 2! = 2*1
### 3! = 3*2*1
### 4! = 4*3*2*1
### 5! = 5*4*3*2*1
### take any number n by user input


```python
num = int(input("Enter number: "))
factorial = 1

for i in range(1, num+1):
    factorial = factorial* i

print("Factorial of", num, "is", factorial)
```
