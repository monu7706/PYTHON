```python
## nested function 
def outer():
    def inner():
        print("i am inside the inner circle")
    inner()
    print("i am outside the inner circle")
outer()
```

```python
## higher outer function
def greet():
    return "Hello ,you'r welcome"
def call_function(func):
    print (func())
call_function(greet)
```

```python
## *args()
def pqr(a,*b):
    print(f"value of a is {a} and value of b is {b}")
pqr(2,4,5,6,7,8)
```

```python
## **kwargs()
def xyz(*a,**b):
    print(f"value of a is {a} and value of b is {b}")
xyz(2,4,5,6,7,8,name="alex,age=45",profession="sde")
```

## local scope and global scope

```python
x = 5
def show():
    print(x)
print(x)
show()
```

```python
x = 5
def show():
    x=9 ## local
    print(x)
print(x)
show()
```

```python
x = 5 ## global
def show():
    global y
    y=9 
    print(x) 
    print(y)
print(x)
show()
print(y)
```
