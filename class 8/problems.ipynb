```python
emails = [
  "alice@example.com",
  "bob@example",
  "charlie@example.com",
  "david@example.com",
  "eve@example.com",
  "frank@example.com",
  "grace@example.com",
  "alice@example.com"]

## EXTRACTS THE NAMES AND PUT IN A LIST .

```

```python
name_list = []
for mail in emails:
    a=mail.split("@")[0]
    name_list.append(a)
print(name_list)
```

```python
## dusra tarika

names = [email.split('@')[0] for email in emails]
print(names)
```
