# Decision Control



## if else


```python
a = 50
b = 60
c = 80
## Q1. if a<b and b<c then print "a<c"
if a<b and b<c:
    print("a<c")
else:
    print("not matching the criteria")
```

## multiple condition with elif

```python
a = 50
b = 60
c = 80
## Q1. if a<b and b<c then print "a<c"
if a<b and b<c:
    print("a<c")
elif a<b or b>c:
    print("only a is less than b but not sure about c ")
else:
    print("not matching the criteria")
```

```python
## take a number(>=0) and check if it is "0" then print "Zero" otherwise "positive number" . 
## if not fall under any of theese then print "Negative number"
```

```python
x = int(input("Enter random number"))
if x>=0:
    if x==0:
        print("Zero")
    else:
        print("Positive number")
else:
    print("negstive number")
```
