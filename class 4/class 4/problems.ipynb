## problem -4 ->Nested Discount Decidion System

```python
## you have to create an online store discount decision logic
## inputs ->
## 1. Cart_value
## 2. is_festival
## 3. is_first_time_customer
## 4.type_of_product
## 5. referral_code
## problem rule:
## basic discount ->
# Cart value<=10000 then provide discount of 5%
# Cart_value<=50000 and >=10000 then provide discount of 10%
# Cart_value>50000 then provide 15% discount 

## if there is festival then provide additional 5% discount for every condition.
## if first time user then provide again additional 5% discount. 
## if product is "electronics" type then provide no additional discount other than
# compulsive offer but if grocery than 2%
## if referral_code ="Mega40%" and minimum cart_value >=55000 then provide flat 40% discount. 
```

```python

cart_value = int(input("Enter cart_value: "))
is_festival = (input("Enter is_festival(yes/no): "))
is_first_time_customer = (input("Enter is_your_first_time(yes/no): "))
type_of_product = (input("Enter type of product(grocery/electronics/others): "))
referral_code = input("Enter referral code: ")

if cart_value<=10000:
    print("discount 5%:", cart_value*0.05)
elif cart_value<=50000 and cart_value >=10000:
    print("discount 10%:", cart_value*0.10)
    if is_festival =="yes":
        print("additional discount 5%: ",cart_value*0.05)
        if is_first_time_customer =="yes":
            print("additional discount 5% again: ",cart_value*0.05)
            if type_of_product == "grocery":
                 print("discount 2%:", cart_value*0.02)
            elif type_of_product == "electronics":
                print("pass")
            else:
                 print("no additional discount provided")
        else:
                print("no additional discount provided")          
                if referral_code == "Mega40%" and cart_value>=55000:
                    print("flat 40% discount: ",cart_value*0.40)
                else:
                    print("no discount provided")
    else:
                print("no additional discount provided")           
else:
    print("discount 15%:", cart_value*0.15)




```
