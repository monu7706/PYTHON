## 4 Data Cleaning

### # Remove duplicates

```python
df.drop_duplicates(inplace=True)
```

### # Fill missing values

```python
df['rating'] = df['rating'].fillna('Not Rated')
df['country'] = df['country'].fillna('Unknown')
df['director'] = df['director'].fillna('Unknown')
df['cast'] = df['cast'].fillna('Unknown')
```

### # Extract numeric duration

```python

df['duration_num'] = df['duration'].str.extract('(\d+)').astype(float)
```

### # Clean type column

```python
df['type'] = df['type'].str.strip()
```

## 6. Basic Visualization (Python)

### 6.1 Movie vs TV Show Distribution 

```python
df['type'].value_counts().plot(kind='pie', autopct='%1.1f%%', figsize=(6,6))
plt.title("Distribution: Movies vs TV Shows")
plt.show()
```

### 6.2 Rating Distribution

```python
plt.figure(figsize=(10,5))
sns.countplot(data=df, x='rating', order=df['rating'].value_counts().index)
plt.xticks(rotation=45)
plt.title("Content Rating Distribution")
plt.show()
```

### 6.3 Release Year Trend

```python
df['release_year'].value_counts().sort_index().plot(kind='line', figsize=(10,5))
plt.title("Content Release Trend Over Years")
plt.xlabel("Year")
plt.ylabel("Number of Titles")
plt.show()
```

### 6.4 Top Genres

```python
plt.figure(figsize=(12,6))
df['listed_in'].value_counts().head(10).plot(kind='bar')
plt.title("Top 10 Genres")
plt.xlabel("Genre")
plt.ylabel("Count")
plt.show()
```
